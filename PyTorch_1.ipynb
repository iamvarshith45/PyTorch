```python
z = torch.tensor([-1, -2, 3, -4, 5])

# absolute value
print(torch.abs(z))

# negative value
print(torch.neg(z))
```

```python
d = torch.tensor([1.74, 2.7, 3.5, 4.56])

# round
print(torch.round(d))

# ceil
print(torch.ceil(d))

# clamp
print(torch.clamp(d, min=2, max=3))
```

```python
m1 = torch.randint(size=(2,3), low=2,high = 10,dtype=torch.float32)
m1
```

```python
# sum
print('sum :',torch.sum(m1))

# sum along columns
print('sum along columns :',torch.sum(m1, dim=0))

# sum along rows
print('sum along rows :',torch.sum(m1, dim=1))
```

```python
# mean
torch.mean(m1)


# mean along col
torch.mean(m1, dim=0)
```

```python
# median
torch.median(m1)

# standard deviation
torch.std(m1)

# variance
torch.var(m1)
```

```python
f = torch.randint(size=(2,2), low=0, high=10)
g = torch.randint(size=(2,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplcation
print('matrix multiplcation :',torch.matmul(f, g))

# transpose
print('transpose :',torch.t(f))

```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
print(h)

print('\n')

#determinant
print('determinant :',torch.det(h))

print('\n')
#inverse
print('inverse :',torch.inverse(h))
```

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)

# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
```

ACTIVATION FUNCTIONS

```python
k = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
k
```

```python
# log
torch.log(k)
```

```python
# sqrt
torch.sqrt(k)
```

```python
# sigmoid
torch.sigmoid(k)
```

```python
# softmax along columns (dim=0)
torch.softmax(k, dim=0)
```

```python
# softmax along rows (dim=1)
torch.softmax(k, dim=1)
```

```python
# relu
torch.relu(k)
```
